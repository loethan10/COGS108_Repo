### EDA: Movie Ratings and Genre Preferences

First, we consider genre versus ratings over time.

```python
genres = cleaned_movies[['genre', 'rating', 'year']].copy()
genres['genre'] = genres['genre'].str.split(', ')
genres = genres.explode('genre')

uniqueGenres = genres[['genre', 'rating', 'year']]

uniqueGenres['genre'].unique()
```

```python
# There are a LOT, so we can combine some and drop some since we only really care about 'Family'
# Define a clean-up method: 
def cleanup(genres_list):
    toDrop = ['Western', 'Sport', 'Mystery', 'Biography', 'History', 'Music', 'Musical' ]
    toReturn = []

    actionAdventure = False
    sciFantasy = False

    for genre in genres_list: 
        if genre in toDrop:
            toReturn.append(np.nan)
        elif genre in ['Action', 'Adventure']:
            actionAdventure = True
        elif genre in ['Sci-Fi', 'Fantasy']:
            sciFantasy = True
        else: 
            toReturn.append(genre)

    if actionAdventure:
        toReturn.append('Action and/or Adventure')
    if sciFantasy:
        toReturn.append('Sci-Fi and/or Fantasy')
    
    return toReturn
```

```python
genres = cleaned_movies[['genre', 'rating', 'year']].copy()
genres['genre'] = genres['genre'].str.split(', ')
genres['genre'] = genres['genre'].apply(cleanup)

genres = genres.explode('genre')
genres = genres.dropna(subset = ['genre'])
genres
```

```python
averageRatings = genres.groupby(['year', 'genre'])['rating'].mean().reset_index()

#Pivot the DataFrame to have years as rows and genres as columns
averageRatings = averageRatings.pivot(index='year', columns='genre', values='rating').fillna(0)
```

```python
#NOW plot
plt.figure(figsize=(12, 6))

# Plot each genre's average rating over the years
for genre in averageRatings.columns:
    plt.plot(averageRatings.index, averageRatings[genre], marker='o', label=genre)

plt.title('Average Rating by Genre Over the Years (2018-2024)')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.xticks(averageRatings.index)  # Set x-ticks to the years
plt.ylim(0, 10)  # Assuming ratings are on a scale of 0 to 10
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Function scrape data from URL
def scrape_box_office(url):

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', class_ = 'mojo-body-table')
    rows = table.find_all('tr')[1:]
    year = url.split('/')[4]
    data = []

    for row in rows:
        cols = row.find_all('td')
        if len(cols) > 7:
            title = cols[1].text.strip()
            gross = cols[7].text.strip()
            data.append({'Title': title, 'Gross': gross, 'Year': year})

    return data

# URLs for scraping
urls = [
    "https://www.boxofficemojo.com/year/2016/?ref_=bo_yl_table_9",
    "https://www.boxofficemojo.com/year/2017/?ref_=bo_yl_table_8",
    "https://www.boxofficemojo.com/year/2018/?ref_=bo_yl_table_7",
    "https://www.boxofficemojo.com/year/2019/?ref_=bo_yl_table_6",
    "https://www.boxofficemojo.com/year/2020/?ref_=bo_yl_table_5",
    "https://www.boxofficemojo.com/year/2021/?ref_=bo_yl_table_4"
]

# Scrape/combine data from URLs
all_data = []
for url in urls:
    all_data.extend(scrape_box_office(url))

# DataFrame from the scraped data
df = pd.DataFrame(all_data)

# Clean/process the 'Total Gross' column
df['Gross'] = df['Gross'].str.replace(r'[\$,]', '', regex = True).astype(float)

# Marvel movies listed by release order for checking (2016-2021)
marvel_movies = [
    "Captain America: Civil War", "Doctor Strange", "Guardians of the Galaxy Vol. 2",
    "Spider-Man: Homecoming", "Thor: Ragnarok", "Black Panther", "Avengers: Infinity War",
    "Ant-Man and the Wasp", "Captain Marvel", "Avengers: Endgame", "Spider-Man: Far from Home",
    "Black Widow", "Shang-Chi and the Legend of the Ten Rings", "Eternals", "Spider-Man: No Way Home"
]

# Filter Marvel movies
filtered_df = df[df['Title'].isin(marvel_movies)].copy()

# Sorting by release order
release_order = {movie: i for i, movie in enumerate(marvel_movies)}
filtered_df['Release Order'] = filtered_df['Title'].map(release_order)
filtered_df.sort_values('Release Order', inplace = True)

# Labels for plot
labels = [f"{row['Title']} ({row['Year']})" for _, row in filtered_df.iterrows()]

# Plotting total gross over time
plt.figure(figsize = (14, 6))
plt.bar(labels, filtered_df['Gross'], color = 'b')
plt.xlabel('Movie Title (Release Year)', fontsize = 12)
plt.ylabel('Total Gross (in billions)', fontsize = 12)
plt.title('Total Gross of Marvel Movies (2016-2021)', fontsize = 14)
plt.xticks(rotation = 45, ha = 'right', fontsize = 8)
plt.tight_layout()
plt.show()

# Checking dataframe
print(filtered_df[['Title', 'Year', 'Gross']])
```

```python
# Fetch webpage
url = "https://www.boxofficemojo.com/year/2020/?ref_=bo_nb_di_secondarytab"
req = requests.get(url)

# Parse webpage with BeautifulSoup
soup = BeautifulSoup(req.content, 'html.parser')

# Locate/extract first table
table = soup.find_all("table")
tbl = table[0]

# Extract rows/skipping the header
data_rows = tbl.find_all('tr')[1:]

# Extract movie details from each row
movie_data = []
for row in data_rows:
    cols = row.find_all('td')
    if len(cols) > 1:
        movie_name = cols[1].get_text().strip() # Movie name
        release_date = cols[7].get_text().strip() # Release date
        total_gross = cols[8].get_text().strip() # Total gross
        movie_data.append([movie_name, release_date, total_gross])

# Create dataFrame for movie data
df_movies = pd.DataFrame(movie_data, columns = ['Movie Name', 'Total Gross', 'Release Date'])

# Clean/preprocess the 'Total Gross' column/remove unecessary data
df_movies['Total Gross'] = df_movies['Total Gross'].replace('-', pd.NA) # Replace '-' with NaN
df_movies['Total Gross'] = df_movies['Total Gross'].str.replace('[\$,]', '', regex = True) # Remove $/commas
df_movies['Total Gross'] = pd.to_numeric(df_movies['Total Gross'], errors = 'coerce') # Convert to numeric

# Extract first three letters of 'Release Date' to represent month
df_movies['Movies'] = df_movies['Release Date'].str[:3]

# Group by month/calculate total revenue
monthly_revenue_2020 = df_movies.groupby('Movies')['Total Gross'].sum()

# Define month order for sorting/make sure all months are included
month_order = {
    "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4,
    "May": 5, "Jun": 6, "Jul": 7, "Aug": 8,
    "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
}
monthly_revenue_2020 = monthly_revenue_2020.reindex(month_order.keys(), fill_value = 0)

# Create dataFrame for plotting
monthly_revenue_2020_df = pd.DataFrame({'Month': monthly_revenue_2020.index, 'Total Revenue': monthly_revenue_2020.values})

# Plotting monthly revenue data
plt.figure(figsize = (10, 6))
plt.plot(monthly_revenue_2020_df['Month'], monthly_revenue_2020_df['Total Revenue'], marker = 'o', linestyle = '-', color = 'b')
plt.title("Monthly Box Office Revenue for 2020")
plt.xlabel("Month")
plt.ylabel("Total Gross (in billions)")
plt.xticks(rotation = 45)
plt.grid(True)

# Display graph
plt.tight_layout()
plt.show()
```

Finally, we want to analyze the year 2020, specifically, to identify whether the shifts (or decline) in revenue coincided with the COVID-19 pandemic lockdowns. As the graph illustrates, there is a clear dip in revenue from February to March; in fact, there is a sharp decline in sales from March all the way to September. Onwards from September, however, the slope maintains a steep positive incline. 

Coming into the year, in January, the COVID-19 pandemic became a paranoid threat amongst the population, as it was introduced in the news. The decline in revenue starting in March directly coincides with the lockdowns, as theater closures started in mid-March. In the time period when the lockdown came into full effect, April to June, the box office markets saw little to no revenue, as movie-goers were unable to go to the theaters. Similarly, therefore closures eased as the lockdown became more relaxed in October, coinciding with the resurgence in sales October and onwards. 

# Ethics & Privacy

Our team aimed to use and collect data only from unbiased and ethical sources, e.g. those without any inherent biases nor those who collected and/or used data with unethical implications. In short, we only used data that we knew we were allowed to and were collected with the users’ consent. 

##### Potential Biases
Early into our research, our team was worried about data from Hollywood causing a skew in our data collection; Hollywood is the heart of cinema and movies, which would most likely cause movie theaters in the area to have a much higher rate of ticket sales than other locations. Furthermore, also early into our research, our team attempted to avoid this bias by limiting our research to just the state of California, but we were unable to find datasets that narrowed down statistics to only our state. Because such information was not readily available, we also did not want to risk using unreliable sources that were not fact checked by either a national institution or a government agency, as they may report biased numbers. As a result, we opened the scope up a little more to the entirety of the United States, as many box offices reported their statistics through the “American Box Office.” We mitigated the biases based towards Hollywood by making sure we had datasets not heavily favoring only the Hollywood area, but rather including a proportionate amount of datasets from other areas; all ticket sale statistics are important, especially possible outliers.  

##### Movie Profits Data and Financial Reports
Several of our box office datasets were pulled from IMDb and other similar sources. In particular, we used a website called The Numbers, which included ticket sale counts reported to the Motion Picture Association of America (MPAA). We were ensured that this data was not biased as, legally, ticket sales must be reported accurately, otherwise specific theaters could be investigated for fraud. As a result, we believe that our data is, to the best of our knowledge, accurate. Further, in regards to legality, tickets are bought with users’ consent, and since the number of tickets bought were reported to a public government agency, identifying and combing through the data was not a breach of privacy. 

##### Consent
As mentioned before, we wanted to ensure that the data we collected wasn’t taken without users’ knowledge. Part of this discussion arose when we debated over whether or not (and how) to include statistics from streaming platforms. However, we decided that, due to the controversy over such data being collected without consent nor knowledge, we omitted such data. We also could not find data that included only those of users in the US, which meant the data would not be applicable to our research.
